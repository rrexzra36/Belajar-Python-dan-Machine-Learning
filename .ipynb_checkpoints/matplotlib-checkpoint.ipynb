```python
# pyplot style
plt.subplot(2,1,1) #(rows, cols, axes 1)
plt.plot(x1, y1, 'ro-')
plt.title('Mutiple Subplots')
plt.ylabel('Nilai $cos(x_1)$')

plt.subplot(2,1,2) #(rows, cols, axes 1)
plt.plot(x2, y2, 'g.-')
plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('Nilai $x$')

plt.show()
```

# 5. Belajar Bar Plot

## 5.1 Simple Bar Plot

```python
data = [25,45,55,125,225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

## 5.2 Pengaturan Grid dan Color

```python
plt.bar(
    kategori,
    data,
    color='red',
    alpha=0.75
)

plt.grid(
    linestyle='--',
    linewidth=2,
    axis='y',
    alpha=0.50
)

plt.xlabel('kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

## 5.3 Horizontal Bar Plot

```python
# plt.barh(kategori, data)
plt.barh(kategori[::-1], data[::-1])

plt.grid(
    linestyle='--',
    linewidth=1,
    axis='x',
    alpha=0.75
)

plt.xlabel('Jumlah')
plt.ylabel('kategori')
plt.title('Contoh Horizontal Bar Plot')
plt.show()
```

## 5.4 Stacked Bar Plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data1, label='Data 1')
# plt.bar(kategori, data2, label='Data 2')
plt.bar (kategori, data2, label='Data 2', bottom=data1)

plt.legend()

plt.grid(
    linestyle='--',
    linewidth=1,
    axis='y',
    alpha=0.75
)
plt.xlabel('kategori')
plt.ylabel('Jumlah')
plt.title('Conton Stacked Bar Plot')

plt.show()
```

## 5.5 Grouped Bar Plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
x = np.arange(len(kategori))
x
```

```python
width = 0.35

plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori)

plt.grid(
    linestyle='--',
    linewidth=1,
    axis='y',
    alpha=0.75
)

plt.xlabel('kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Grouped Bar Plot')

plt.legend()
plt.show()
```

# 9. Matplotlib Style

## 9.1 Fungsi Plotting

```python
def gambar_barplot():
    data1 = [25,85,75,40,60]
    data2 = [40,35,20,55,10]
    kategori = ['A', 'B', 'C', 'D', 'E']

    x = np.arange(len(kategori))

    width = 0.35

    plt.bar(x-width/2, data1, width, label='Data 1')
    plt.bar(x+width/2, data2, width, label='Data 2')

    plt.xticks(x, kategori)

    plt.grid(
        linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75
    )

    plt.xlabel('Kategori')
    plt.ylabel('Jumlah')
    plt.title('Contoh Grouped Bar Plot')

    plt.legend()
    plt.show()
```

# 13. Twin Axes

## 13.1 Kasus Twin Axes 1

```python
x = np.linspace(1, 10, 25)
x
```

```python
# OO style
fig, ax1 = plt.subplots()

ax1.plot(x, np.exp(x), 'bs-', label='exp')
ax1.set_xlabel('Sumbu x')
ax1.set_ylabel('Sumbu y')

ax2 = ax1.twinx() # menghasilkan axes baru bertumpukan dengan axes yang pertama
ax2.plot(x, np.log(x), 'ro-', label='log')
ax2.set_ylabel('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc='upper left')

plt.show()
```

## 12.2 Kasus twin axes 2

```python
x = np.random.randint(16, 40, size=30)
x
```

```python
def C2F(celcius=0):
    return (celcius*1.8)+32

def konversi_sumbu(ax1):
    y1, y2 = ax1.get_ylim()
    ax2.set_ylim(C2F(y1), C2F(y2))
    ax2.figure.canvas.draw()

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.callbacks.connect('ylim_changed', konversi_sumbu)
ax1.plot(x)

ax1.set_xlabel('Hari')
ax1.set_ylabel('Celcius')
ax2.set_ylabel('Fahrenheit')


fig.suptitle('Temperatur Udara')
plt.show()
```

## 16.1 Sample Dataset

```python
items = tuple(data.keys())
items
```

```python
count = tuple(data.values())
count
```

```python
fig, ax = plt.subplots()

ax.barh(items, count)

plt.show()
```

## 16.3 Pengaturan style

```python
plt.style.available
```

## 16.4 Pengaturan tick label

```python
fig, ax = plt.subplots()
ax.barh(items, count)

labels = ax.get_xticklabels()

plt.setp(
    labels,
    rotation=45,
    horizontalalignment='right'
)

plt.show()
```

## 16.4 Pengturan format pada ticker

```python
from matplotlib.ticker import FuncFormatter
```

```python
def ribuan(x, pos):
    return f'{int(x/1000)}K'
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)

ax.set(
    title='Contoh Plot',
    xlabel='Jumlah',
    ylabel='items'
)

plt.show()
```

## 16.5 Pengaturan label pada sumbu (axis) dan judul (title)

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.set(
    title='Contoh Plot',
    xlabel='Jumlah',
    ylabel='items'
)

plt.show()
```

## 16.6 Penambahan garis (vert/ horizon)

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count),
           ls='--',
           color='r')

ax.set(
    title='Contoh Plot',
    xlabel='Jumlah',
    ylabel='items'
)

plt.show()
```

# 19. Pengaturan Figure Layouts Dengan Grid Specs

## 19.1 Simple 2x2 Layout dengan suplots

```python
siswa = ['Cecep', 'Joko', 'Wati', 'Ani']
uas = [68, 83, 96, 72]

fig, axes = plt.subplots(ncols=2, nrows=2, constrained_layout=True)

axes[0][0].bar(siswa, uas)
axes[0][1].pie(uas, labels=siswa)
axes[1][0].pie(uas, labels=siswa, explode=[0,0,0,0.2])
axes[1][1].barh(siswa, uas)

plt.show()
```

## 19.2 Simple 2x3 layout dengan gridspec

```python
fig = plt.figure(constrained_layout=True)

gs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.bar(siswa,uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0,0,0,0.2])
ax4.barh(siswa, uas)

plt.show()
```

## 18.3 Rows dan cols span dengan gridspec

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(ncols=2, nrows=2)

ax1 = fig.add_subplot(gs[0, :])
#ax2 = fig.add_subplot(gs [0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

ax1.bar(siswa, uas)
#ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode= [0, 0, 0, 0.2])
ax4.barh(siswa, uas)
plt.show()
```

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(ncols=3, nrows=3)

ax1 = fig.add_subplot(gs[0, :])
ax1.set_title('gs[0, :]')

ax2 = fig.add_subplot(gs[1, :-1])
ax2.set_title('gs[1, -1]')
ax2.barh(siswa, uas)

ax3 = fig.add_subplot(gs[1:, -1])
ax3.set_title('gs[1:, -1]')
ax3.bar(siswa, uas)

ax4 = fig.add_subplot(gs[-1, 0])
ax4.set_title('gs[-1, 0]')

ax5 = fig.add_subplot(gs [-1, -2])
ax5.set_title('gs[-1, -2]')

plt.show()
```

# 19. Mengenal Animasi

```python
from matplotlib.animation import FuncAnimation
from IPython import display
```

## 19.1 Review Figure

```python
fig = plt.figure()

plt.xlim((0, 4))
plt.ylim((-2, 2))

plt.plot([])
```

## 19.2 Animasi pada matplotlib

```python
fig = plt.figure()
plt.xlim((0,4))
plt.ylim((-2,2))

line = plt.plot([])[0]

def animate(i):
    x = np.linspace(0, 4, 100)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x,y)
    return line,

anim = FuncAnimation(fig, animate, frames=200, interval=20)

video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)

plt.close()
```
